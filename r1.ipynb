```python
import math
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
Data = pd.read_csv("C:\\Users\\Deepa\\Desktop\\pracs\\bank.csv")

print(Data)
```

```python
cols =["age","balance","day","campaign","pdays","previous"]
data_encode=Data.drop(cols,axis=1).apply(LabelEncoder().fit_transform)
data_rest=Data[cols]
print(data_rest)
Data=pd.concat([data_rest,data_encode],axis=1)
print(Data)
```

```python
X=Data.drop("Target",axis=1)
y=Data["Target"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.6,random_state=4)
```

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
Kcents=8
km=KMeans(n_clusters=Kcents,max_iter=1000,random_state=42)
km.fit(X_train)
centers=km.cluster_centers_
```

```python
#calculate sigma

max_dist=0.0
for i in range(Kcents):
    for j in range(Kcents):
        d=np.linalg.norm(centers[i]-centers[j])
        if d > max_dist:
           max_dist=d
sigma=max_dist/math.sqrt(2*Kcents)            
```

```python
#gaussian matrix
#create an empty 2d numpy array

G_train=np.empty((X_train.shape[0],Kcents),dtype=float)
for i in range(X_train.shape[0]):
    for j in range(Kcents):
        dist=np.linalg.norm(X_train[i]-centers[j])
        G_train [i][j]=math.exp(-math.pow(dist,2)/math.pow(2*sigma,2))
        
G_test = np.empty((X_test.shape[0], Kcents), dtype=float)
for i in range(X_test.shape[0]):
    for j in range(Kcents):
        dist = np.linalg.norm(X_test[i] - centers[j])
        G_test[i][j] = math.exp(-math.pow(dist, 2) / math.pow(2 * sigma, 2))
    
```

```python
#weight matrix
#W=(G ^T. G) ^−1. G ^T .Y

GTG=np.dot(G_train.T,G_train)
GTG_inv=np.linalg.inv(GTG)
res=np.dot(GTG_inv,G_train.T)
W=np.dot(res,y_train)
```

```python
#predict
#GW=Y

predict = np.dot(G_test, W)
predict = 0.5 * (np.sign(predict - 0.5) + 1)  # Converts to 0 or 1

score = accuracy_score(predict, y_test)
print("\nAccuracy: ", score)
```

```python
#wine dataset

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import math
```

```python
Data1=pd.read_csv("C:\\Users\\Deepa\\Desktop\\pracs\\wine_data.csv")
print(Data1)
```

```python
# Convert quality to 3-class label: 0 (low), 1 (medium), 2 (high)
Data1["quality"] = Data1["quality"].apply(lambda q: 0 if q <= 5 else 1 if q == 6 else 2)

```

```python
print(Data1)
```

```python
X=Data1.drop("quality",axis=1)
y=Data1["quality"]

print(y)
print(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.6,random_state=4)
```

```python
#normalise my data

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#centroid calc
K_cents=12
km=KMeans(n_clusters=K_cents,max_iter=1000,random_state=4)
km.fit(X_train)
cent=km.cluster_centers_
```

```python
#sigma calc
maxdist=0.0
for i in range(K_cents):
    for j in range(K_cents):
        d=np.linalg.norm(cent[i]-cent[j])
        if d > maxdist:
            maxdist=d
sigma=maxdist/math.sqrt(2*K_cents)
```

```python
#gaussian train and test calc

G1_Train=np.empty((X_train.shape[0],K_cents),dtype=float)
for i in range(X_train.shape[0]):
    for j in range(K_cents):
        d1=np.linalg.norm(X_train[i]-cent[j])
        G1_Train[i][j]=math.exp(-math.pow(d1,2)/math.pow(2*sigma,2))

G1_Test=np.empty((X_test.shape[0],K_cents),dtype=float)
for i in range(X_test.shape[0]):
    for j in range(K_cents):
        d1=np.linalg.norm(X_test[i]-cent[j])
        G1_Test[i][j]=math.exp(-math.pow(d1,2)/math.pow(2*sigma,2))
        
```

```python
#weight calc

GTG1=np.dot(G1_Train.T,G1_Train)
GTG1_inv=np.linalg.inv(GTG1)
res=np.dot(GTG1_inv,G1_Train.T)
W1=np.dot(res,y_train)
```

```python
#predict G.W=Y
predict = np.dot(G1_Test, W1)
predict = np.round(predict).astype(int)

# Clip predictions to stay in [0, 2]
predict = np.clip(predict, 0, 2)

# Accuracy
score = accuracy_score(y_test, predict)
print("\nImproved RBF Accuracy (3-class):", round(score * 100, 2), "%")
```

```python
#final
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

x, y = load_wine(return_X_y = True)

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

def rbf_kernel(x1, x2, gamma):
    diff = x1 - x2
    return np.exp(-gamma * np.dot(diff, diff.T))

def rbf_classifier(x_train, y_train, x_test, gamma):
    predictions = []

    for test_sample in x_test:
        distances = []

        for train_sample in x_train:
            distance = rbf_kernel(test_sample, train_sample, gamma)
            distances.append(distance)

        closest_index = np.argmax(distances)
        predictions.append(y_train[closest_index])

    return np.array(predictions)

predictions = rbf_classifier(x_train, y_train, x_test, gamma=0.5)
print(predictions)
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from collections import Counter

# Load dataset
df = pd.read_csv("C:\\Users\\Deepa\\Desktop\\pracs\\wine_data.csv")

# Features and target
X = df.drop('quality', axis=1)
y = df['quality']

# Reclassify quality into 3 classes: low (0), medium (1), high (2)
y = y.apply(lambda q: 0 if q <= 5 else 1 if q == 6 else 2)

# Check class distribution
print("Class distribution:", Counter(y))

# One-hot encode target
y_cat = to_categorical(y, num_classes=3)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42, stratify=y)

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build improved ANN
model = Sequential()
model.add(Dense(256, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))  # Regularization
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dense(3, activation='softmax'))  # Output layer

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train model
history = model.fit(
    X_train, y_train,
    epochs=150,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate on test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f'\nTest Accuracy: {accuracy * 100:.2f}%')

# Predict and print classification report
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

print("\nClassification Report:\n", classification_report(y_true, y_pred_classes))
```
